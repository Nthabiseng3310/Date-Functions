```sql
-- The function called CURRENT_DATE() gives us today's date
SELECT CURRENT_DATE() AS today_dt;

-- The function called CURRENT_TIMESTAMP gives me the current date and time
SELECT CURRENT_TIMESTAMP() AS dt_time;

-- Converts a STRING(words) into a date (Ensure that a date is read as a date)
SELECT TO_DATE('2026-07-17') AS dt;

-- Converts a date into a string
SELECT TO_CHAR(TO_DATE('2026-07-17'),'yyyyMM') AS  Month_id; 
SELECT TO_CHAR(TO_DATE('2026-07-17'),'yyyy') AS  year_value; 
SELECT TO_CHAR(TO_DATE('2026-07-17'),'dd') AS  day_value; 
SELECT TO_CHAR(TO_DATE('2026-07-17'),'MM') AS  Month_value; 

-- The system recognizes the date format in this way 'yyyy-MM-dd'

-- DATEDIFF(end_date,start_date)
-- end_date - start_date = value 
SELECT DATEDIFF('2026-07-05','2026-07-01') AS dt_difference;

-- DATE_ADD(date, number_of_days)
SELECT DATE_ADD('2026-07-05',5 ) AS add_5_days;
SELECT DATE_ADD('2026-07-05',-5) As sub_5_days;--subtracting 5 days

-- DATE_SUB(date, number_of_days)
SELECT DATE_SUB('2026-07-05', 2) AS removing_2_day;

SELECT TO_CHAR(TO_DATE('2026-07-05'),'MM') AS month_value;
SELECT TO_CHAR(TO_DATE('2026-07-05'),'DD') AS day_value;

SELECT YEAR('2026-07-05') AS year_value;--Extracts year from date
SELECT MONTH('2026-07-05')AS month_dt;--Extracts month from date
SELECT DAY('2026-07-05')AS day_dt;--Extracts day from date

SELECT DAYNAME('2026-07-05')AS day_name;--It gives out the name of the day
SELECT MONTHNAME('2026-07-05')AS month_name;--It gives out the name of the month


SELECT YEAR('2026-07-05') AS year_name;

-------------------------------------------------------------------------------

SELECT DAYOFWEEK('2026-07-05')AS day_name;--Giving out the day of the week
SELECT ADD_MONTHS('2026-07-05',2) AS adding_2_month;--Adding 2 months to date
SELECT LAST_DAY('2026-07-05') AS last_day;--giving the last day of the month
```
